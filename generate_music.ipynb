```python
import torchaudio
from audiocraft.models import MusicGen
from audiocraft.data.audio import audio_write
from torchaudio.transforms import Resample

model = MusicGen.get_pretrained('facebook/musicgen-melody-large')
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
import json
import cv2

file_name = 'city'
with open(f"/content/drive/MyDrive/Video2Music/{file_name}_generation_info.json", "r") as infile:
    loaded_dict = json.load(infile)
print(loaded_dict)

# video time extraction
file_name = loaded_dict['file_name']
video_path = f"/content/drive/MyDrive/Video2Music/{file_name}.mp4"
video = cv2.VideoCapture(video_path)
total_frames = int(video.get(cv2.CAP_PROP_FRAME_COUNT))

fps = video.get(cv2.CAP_PROP_FPS)
video_length = total_frames / fps # all frames / fps = video length
video.release()

print(f"video length: {video_length:.2f}sec")
```

```python
model.set_generation_params(duration=video_length,  # generate for video_length
                            cfg_coef=3.0,
                            # top_k=500,
                            # top_p=0.95,
                            # temperature=0.5,
                            use_sampling=True)

descriptions = [
    loaded_dict['describe_text']
]

melody, orig_sr = torchaudio.load('/content/drive/MyDrive/Video2Music/wav_output/output_with_first_scale.wav')  # output_with_randomness.wav / output_with_first_scale.wav
classified_note = loaded_dict['classified_note']

resample_transform = Resample(orig_sr, 32000)
melody_resampled = resample_transform(melody)

wav = model.generate_with_chroma(descriptions, melody_resampled[None].expand(len(descriptions), -1, -1), 32000)

for idx, one_wav in enumerate(wav):
    audio_write(f'/content/drive/MyDrive/Video2Music/music_output/{file_name}_{classified_note}_2', one_wav.cpu(), model.sample_rate, strategy="loudness")  # strategy="loudness"

print('music generated!')
```
