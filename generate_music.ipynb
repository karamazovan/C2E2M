```python
import torchaudio
from audiocraft.models import MusicGen
from audiocraft.data.audio import audio_write
from torchaudio.transforms import Resample

model = MusicGen.get_pretrained('facebook/musicgen-melody')
model.set_generation_params(duration=16,  # generate 8 seconds.
                            cfg_coef=3.0,
                            # top_k=500,
                            # top_p=0.95,
                            # temperature=0.5,
                            use_sampling=True)
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
import json

file_name = '바다'
with open(f"/content/drive/MyDrive/Video2Music/{file_name}_generation_info.json", "r") as infile:
    loaded_dict = json.load(infile)

print(loaded_dict)
```

```python
descriptions = [
    loaded_dict['describe_text']
]

melody, orig_sr = torchaudio.load('/content/drive/MyDrive/Video2Music/wav_output/output.wav')  # output_with_randomness.wav
file_name = loaded_dict['file_name']
classified_note = loaded_dict['classified_note']

resample_transform = Resample(orig_sr, 32000)
melody_resampled = resample_transform(melody)

wav = model.generate_with_chroma(descriptions, melody_resampled[None].expand(len(descriptions), -1, -1), 32000)

for idx, one_wav in enumerate(wav):
    audio_write(f'/content/drive/MyDrive/Video2Music/music_output/{file_name}_{classified_note}_{idx}', one_wav.cpu(), model.sample_rate, strategy="loudness")  # strategy="loudness"

print('music generated!')
```

```python
from IPython.display import Audio

Audio(wav[0].detach().cpu().numpy(), rate=32000)
```

```python
from IPython.display import HTML
from base64 import b64encode

video_path = f"/content/drive/MyDrive/Video2Music/{file_name}.mp4"

with open(video_path, 'rb') as f:
    video_data = f.read()
    video_encoded = b64encode(video_data).decode('ascii')

HTML(f"""
<video width="640" height="480" controls>
    <source src="data:video/mp4;base64,{video_encoded}" type="video/mp4">
</video>
""")
```
